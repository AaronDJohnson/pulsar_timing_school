```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from PAL2 import bayesutils as bu
import scipy.special as ss
import sys

from simple_mcmc import SimpleMCMC

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python

class SimpleMCMC(object):
    
    def __init__(self, lnlikefn, lnpriorfn, sigmas):
        """
        Simple MCMC Sampler class that will perform
        sampling using gaussian jump proposals
        
        :param lnlikefun: ln-likelihood function.
        :param lnpriorfun: ln-prior function.
        :param sigmas: jump size (either single value or vector size of parameter array)
        
        """
        
        # collect input values
        self.lnlikefun = lnlikefn
        self.lnpriorfn = lnpriorfn
        self.sigmas = np.atleast_1d(sigmas)
        
    
    def sample(self, p0, N):
        """
        Sample for N steps.
        
        :param p0: Initial parameter array
        :param N: Number of iterations
        
        """
        
        ndim = len(p0)
        naccepted = 0
        
        # initialize arrays
        self.chain = np.zeros((N, ndim))
        self.lnlike, self.lnprob = np.zeros(N), np.zeros(N)
        
        # get likelihood and prior for initial point
        lnlike0, lnprior0 = self.lnlikefun(p0), self.lnpriorfn(p0)
        
        it = 0
        # start iterations
        for ii in range(N):
            
            # jump
            p1, qxy = self.jump(p0)
            
            # compute likelihoods and priors for new point
            lnlike1, lnprior1 = self.lnlikefun(p1), self.lnpriorfn(p1)
            
            # hastings step
            diff = (lnlike1 + lnprior1) - (lnlike0 + lnprior0) + qxy
            if diff > np.log(np.random.rand()):

                # accept jump
                p0, lnlike0, lnprior0 = p1, lnlike1, lnprior1
                naccepted += 1
                
            # update arrays
            self.chain[ii,:] = p0
            self.lnlike[ii], self.lnprob[ii] = lnlike0, (lnlike0 + lnprior0)

            
            if it % 100 == 0 and it > 0:
                sys.stdout.write('\r')
                sys.stdout.write('Finished %g percent. Acceptance Rate: %g'%(it / N * 100, naccepted / it * 100))
                sys.stdout.flush()
            
            it += 1
            
    
    def jump(self, p0):
        
        """
        Standard gaussian jump
        
        :param p0: Current parameter vector
        
        :returns: new parameter vector, transition probability
        """
        
        qxy = 0
        q = p0.copy()
        
        q += np.random.randn(len(q)) * self.sigmas
        
        return q, qxy
```

```python
def get_acf(x):
    
    N = len(x)
    sig = np.std(x)
    mu = np.mean(x)
    
    ret = np.zeros(N)
    for h in range(N):
        
        ret[h] = 1 / (N - h) * np.sum((x[h:N]-mu) * (x[:(N-h)]-mu)) / sig**2
        
    return ret
        
```

```python
def lnlikefn(x):
    
    return -0.5 * np.sum(x**2/2)

def lnpriorfn(x):
    
    return 0
```

```python
ndim = 5
N = 10000
p0 = np.random.randn(ndim)
sigmas = 1.0

sampler = SimpleMCMC(lnlikefn, lnpriorfn, sigmas)
sampler.sample(p0, N)
```

## Exercise 3: Scaled/Shifted student-t distribution

```python
class StudentTDistribution(object):
    
    def __init__(self, nu, sigma, mu):
        
        self.nu = nu
        self.sigma = sigma
        self.mu = mu
        
    def lnlikefn(self, x):
        
        prefac = ss.gamma((self.nu + 1) / 2) / ss.gamma(self.nu / 2) / np.sqrt(self.nu * np.pi) / self.sigma
        fac = (1 + 1 / self.nu * ((x - self.mu) / self.sigma)**2)**(-(self.nu + 1) / 2)
        
        return np.log(fac*prefac)
    
    def lnpriorfn(self, x):
        
        return 0
```

```python
dst = StudentTDistribution(3, 1, 1)
x = np.linspace(-8, 8, 1000)
y = np.exp(dst.lnlikefn(x))

plt.plot(x, y)
```

### Small Step Size

```python
# initial parameters
ndim = 1
N = 100000
p0 = np.random.randn(ndim)

sigmas = 0.01
sampler = SimpleMCMC(dst.lnlikefn, dst.lnpriorfn, sigmas)
sampler.sample(p0, N)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(131)
plt.plot(sampler.chain[:])

plt.subplot(132)
a = plt.hist(sampler.chain[:], 50, normed=True)
plt.plot(x, y, lw=2, color='r')
plt.xlim(-6, 8)

plt.subplot(133)
acf = get_acf(sampler.chain[::5,0])
plt.plot(acf)

acl = np.flatnonzero(acf <= 0.37)[0]
print 'ACL = {0}'.format(acl * 5)
```
