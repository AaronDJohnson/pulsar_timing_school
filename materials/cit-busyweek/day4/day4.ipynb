# Exercise 1b #
Check whether the function to create the Hellings&Downs correlation matrix is correct. The H&D curve is given by:

$x_{ij} = \frac{(1-\cos \theta_{ij})}{2}, \quad y_{ij} = \frac{3}{2} x_{ij} \log x_{ij}$

$\Gamma_{ij} = y_{ij} - \frac{1}{4}x_{ij} + \frac{1}{2}(1+\delta_{ij})$

Here $\Gamma$ is the correlation between two pulsars, and $\theta_{ij}$ is the angular separation between the pulsar $i$ and $j$.

_hint:_ Somewhere in the function 'hdcormat' below, it should be possible somewhere to calculate the angular separation between two pulsars.

```python
# Construct the Hellings & Downs correlation matrix
def hdcormat(psrs):
    """ Constructs a correlation matrix consisting of the Hellings & Downs
        correlation coefficients. See Eq. (A30) of Lee, Jenet, and
        Price ApJ 684:1304 (2008) for details.

        :param psrs:
            list of Pulsar objects
        
        Note: the RAJ and DECJ are assumed to reside in psrs[i].raj and psrs[i].decj
        
    """
    # Create the pulsar unit coordinates (pp), pointing towards the pulsars
    npsrs = len(psrs)
    raj = [psr.raj for psr in psrs]
    decj = [psr.decj for psr in psrs]
    pp = np.array([np.cos(decj)*np.cos(raj), np.cos(decj)*np.sin(raj), np.sin(decj)]).T
    
    # cosp is the inner-product between the two pulsar unit coordinates
    cosp = np.array([[np.dot(pp[i], pp[j]) for i in range(npsrs)] for j in range(npsrs)])
    cosp[cosp > 1.0] = 1.0
    xp = 0.5 * (1 - cosp)

    # Do some numpy magic to avoid errors of log(0)
    old_settings = np.seterr(all='ignore')
    yp = 1.5 * xp * np.log(xp)
    np.fill_diagonal(yp, 0)
    np.seterr(**old_settings)
    hdmat = yp - 0.25 * xp + 0.5 + 0.5 * np.diag(np.ones(len(psrs)))

    return hdmat
```

In this exercise, you will finalize the rank-reduced likelihood function for the whole PTA. As a reference, you are given the full time-domain variant, which until 2012 was the only way we were really doing it. You'll see that 36 pulsars would get rather slow, even in the rank-reduced formalism, so use a number that makes the calculations go (somewhat) quickly.

```python
def pl_time_ptaloglikelihood(psrs, efacs, Amps, Sis, gwAmp, gwSi):
    """
    Log-likelihood for our pulsar (time-domain)

    :param psr:
        pulsar object, containing the data and stuff

    :param efacs:
        Vector of efacs
    
    :param Amps:
        Vector of red noise amplitudes
    
    :param Sis:
        Vector of red noise spectral indices
    
    :param gwAmp:
        GW power-law amplitude
    
    :param gwSi:
        GW power-law spectral index
    """
    Nvecs = [psrs[ii].toaerrs**2 * efacs[ii]**2 for ii in range(len(psrs))]
    Mmat = sl.block_diag(*[psr.Mmat for psr in psrs])

    hdmat = hdcormat(psrs)

    residuals = np.hstack([psr.residuals for psr in psrs])
    toas = np.hstack([psr.toas for psr in psrs])
    n = len(toas)

    # Create the full covariance matrix, as a single power-law signal
    Cov = PL_covmat(toas, gwAmp, 0.5*(3-gwSi), fL=1.0/(year*20))

    # Multiply the matrix with the H&D coefficients
    indexa, indexb = 0, 0
    for aa, psra in enumerate(psrs):
        ntoasa = len(psra.toas)

        for bb, psrb in enumerate(psrs):
            ntoasb = len(psrb.toas)

            # Multiply with the H&D coefficient
            Cov[indexa:indexa+ntoasa,indexb:indexb+ntoasb] *= hdmat[aa,bb]

            indexb += ntoasb

        indexb = 0
        indexa += ntoasa

    # Add the white noise
    Cov += np.diag(np.hstack(Nvecs))

    # Add the individual red noise and the GWB
    offset = 0
    for aa, psr in enumerate(psrs):
        ntoas = len(psr.toas)
        Cov[offset:offset+ntoas,offset:offset+ntoas] += \
                PL_covmat(psr.toas, Amps[aa], 0.5*(3-Sis[aa]), fL=1.0/(year*20))
        offset += ntoas

    # From here on, we do the same linear algebra as we did for a single pulsar
    
    # Obtain the Cholesky factor, and thus also the determinant
    cfC = sl.cho_factor(Cov)
    Cinv = sl.cho_solve(cfC, np.eye(len(toas)))
    ldetC = 2 * np.sum(np.log(np.diag(cfC[0])))

    # Calculate M.T C^{-1} M, and get it's Cholesky factor
    MCM = np.dot(Mmat.T, np.dot(Cinv, Mmat))
    cfM = sl.cho_factor(MCM)
    ldetM = 2 * np.sum(np.log(np.diag(cfM[0])))
    
    # Collapse the matrices to vectors for speedy evaluation
    wr = np.dot(Cinv, residuals)
    rCr = np.dot(residuals, wr)
    MCr = np.dot(Mmat.T, wr)
    
    return -0.5 * rCr + 0.5 * np.dot(MCr, sl.cho_solve(cfM, MCr)) - \
            0.5 * ldetC - 0.5 * ldetM -0.5*n*np.log(2*np.pi)
```

```python
def ptaloglikelihood(psrs, Nvecs, psds, gwpsd):
    """
    Log-likelihood for our pulsar (hierarchical/Woodbury)

    :param psr:
        pulsar object, containing the data and stuff

    :param Nvec:
        Vector of squared white noise amplitude (diagonal of covariance matrix)

    :param psd:
        Power-spectral density per frequency bin (of size 1/T)
    
    Assume all pulsars use the same number of frequencies
    """
    residuals = np.hstack([psr.residuals for psr in psrs])
    Nvec = np.hstack(Nvecs)
    
    # Create the full M-matrix, the full F-matrix, and the full T-matrix.
    # The full T-matrix is just (M F), where M and F are the block-diagonal
    # collections of all M and F matrices. So the full T-matrix is structured
    # as follows:
    #       M1  0  0 .. F1  0  0 ..
    #        0 M2  0 ..  0 F2  0 ..
    #        0  0 M3 ..  0  0 F3 ..
    #        .  .  .  .  .  .  .  .
    # 
    Mmat = sl.block_diag(*[psr.Mmat for psr in psrs])
    Fmat = sl.block_diag(*[psr.Fmat for psr in psrs])
    Tmat = np.append(Mmat, Fmat, axis=1)
        
    # Create the T.T N^{-1} T block-matrix
    # Try to speed this up if you can. This is where the majority of the
    # computational time is spent at the moment.
    TNT = np.dot(Tmat.T / Nvec, Tmat)
    
    # Create the H&D matrix
    hdmat = hdcormat(psrs)
    
    ##############################################################
    # For Excersize 2b:
    # Create the phi-matrix here, including the H&D correlations
    # The phi-matrix only works as a prior on Fmat. Notice below
    # how we use an offset when we add Phi_inv to Sigma_inv:
    # Sigma_inv[offset:,offset:] += Phi_inv
    # This offset makes it skip all the design matrix (Mmat) terms
    ##############################################################
    
    # Invert Phi
    cf_phi = sl.cho_factor(Phi)
    Phi_inv = sl.cho_solve(cf_phi, np.eye(len(Phi)))
    
    # Create Sigma_inv and Sigma
    offset = Mmat.shape[1]
    Sigma_inv = TNT.copy()
    Sigma_inv[offset:,offset:] += Phi_inv
    cf_sigmai = sl.cho_factor(Sigma_inv)
    
    # The log-determinants (using Sylvester)
    ldN = np.sum(np.log(Nvec))
    ldP = 2*np.sum(np.log(np.diag(cf_phi[0])))
    ldS = 2*np.sum(np.log(np.diag(cf_sigmai[0])))
    
    # The Woodbury inner-product
    TNr = np.dot(Tmat.T, residuals / Nvec)
    innerprod = np.sum(residuals**2/Nvec) - np.dot(TNr, sl.cho_solve(cf_sigmai, TNr))
    
    return -0.5*innerprod - 0.5*(ldN+ldP+ldS) - 0.5*len(residuals)*np.log(2*np.pi)

def pl_ptaloglikelihood(psrs, efacs, Amps, Sis, gwAmp, gwSi):
    """
    Log-likelihood for our pulsar (hierarchical/Woodbury)

    :param psr:
        pulsar object, containing the data and stuff

    :param efacs:
        Vector of efacs
    
    :param Amps:
        Vector of red noise amplitudes
    
    :param Sis:
        Vector of red noise spectral indices
    
    :param gwAmp:
        GW power-law amplitude
    
    :param gwSi:
        GW power-law spectral index
    
    Assume all pulsars use the same number of frequencies
    """
    Nvecs = [psrs[ii].toaerrs**2 * efacs[ii]**2 for ii in range(len(psrs))]
    psds = [PL_psd(psrs[ii].freqs, Amps[ii], 0.5*(3-Sis[ii]), sig_fL=1.0/(20.0*year)) for ii in range(len(psrs))]
    gwpsd = PL_psd(psrs[0].freqs, gwAmp, 0.5*(3-gwSi), sig_fL=1.0/(20.0*year))
    
    return ptaloglikelihood(psrs, Nvecs, psds, gwpsd)
```
